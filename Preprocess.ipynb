```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data/rawGlobalCO2.csv')

df.drop(0, inplace=True)


rename = {
    'Total carbon emissions from fossil fuel consumption and cement production (million metric tons of C)':'Year Total',
    'Carbon emissions from gas fuel consumption':'Gas Fuel',
    'Carbon emissions from liquid fuel consumption': 'Liquid Fuel',
    'Carbon emissions from solid fuel consumption': 'Solid Fuel',
    'Carbon emissions from cement production': 'Cement', #emission from cement production
    'Carbon emissions from gas flaring': 'Gas Flaring', #emission from gas flaring 
    'Per capita carbon emissions (metric tons of carbon; after 1949 only)': 'Per Capita'
}
df.rename(columns=rename, inplace=True)

for col in df:
    if col == 'Per Capita':
        continue
    df[col] = df[col].astype(int)
    
df = df[df['Year'] >= 1900]
df.set_index('Year', inplace=True)

for year in df.index:
    df.loc[year, 'Year Total'] = df.loc[year, 'Gas Fuel'] + df.loc[year, 'Liquid Fuel'] + df.loc[year, 'Solid Fuel']

for year in df.index:
    df.loc[year, 'Total'] = sum(df.loc[:year, 'Year Total'])
    
df.head(10)
```

```python
df.to_csv('Data/GlobalCO2.csv')
```
